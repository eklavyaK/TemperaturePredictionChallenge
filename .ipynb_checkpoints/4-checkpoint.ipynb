```python
import numpy as np
import tensorflow as tf
import csv
import warnings
```

#### ***Loading the dataset***

```python
obj = csv.reader(open("train_x.csv", 'r'))
DATA = np.array(list(obj))
print(DATA.shape)
```

Since csv reader reads the dataset in string format, we need to convert the cells in their respective formats.
The 1st and 2nd column of every hour is in form of string, so we'll use number encoding to train them.

```python
summary, PrecipType, others = set(), set(), set()
for i, data in enumerate(DATA):
	if i == 0: continue
	for j, entry in enumerate(data):
		if j % 7 == 0: summary.add(entry) # creating a set of 1st and 2nd feature of every hour
		elif j % 7 == 1: PrecipType.add(entry)
		elif(entry.replace('.', '', 1).replace('-', '', 1).isdigit() == False): others.add(entry) # creating a set of the all the floating point entries to make recognize the entries which are non - floating point and replace with appropriate floating point numeral
print(len(summary), len(PrecipType), len(others))
print(summary)
print(PrecipType)
print(others)
```

We got 26 'summary' feature types. These are strings we'll encode them from 0 to 25. And Precipitation feature can take 3 different values, where one of values is unknown, we'll replace it with 0.

```python
summarydict, precipdict, code = {}, {}, -1
for i in summary:  # forming dictionary to map strings with their code
	code += 1
	summarydict[i] = code
code = -1
for i in PrecipType:
	code += 1
	precipdict[i] = code

print(summarydict)
print(precipdict)
```

We've create a dictionary to map the strings with their respective code. Now we'll create the final dataset by converting the strings to floating point numbers.

```python
for i, data in enumerate(DATA):
	if i == 0: continue
	for j, entry in enumerate(data):
		if j % 7 == 0: DATA[i, j] = summarydict[entry]    # mapping the strings with codes
		elif j % 7 == 1: DATA[i, j] = precipdict[entry]
		else: DATA[i, j] = float(entry)      # typecasting string numbers to floating point numerals
for i in range(1, 6):
	print(DATA[i])
```

We've total 77158 records in the dataset. Now we'll divide the dataset into training, validation and test sets in following way:
<br>
<Br>
*Total records in training set*: **53899** <br>
*Total records in validation set*: **11640** <br>
*Total records in test set*: **11619** <br>

<br>
RFtest data are created to be used for training different type of model where validation data is not required.

```python
train_x, dev_x, test_x = DATA[1:53900].astype(float), DATA[53900:65540].astype(float), DATA[65540:].astype(float) # data split
RFtrain_x, RFtest_x = DATA[1:65000].astype(float), DATA[65000:].astype(float)
print(train_x.shape, dev_x.shape, test_x.shape)
for i in range(5):
	print(train_x[i])
```

##### ***Processing the Targets***

```python
warnings.filterwarnings('ignore')
obj = csv.reader(open("train_y.csv", 'r'))
DATA = list(obj)
print(len(DATA))
for i in range(1, 6):
	print(DATA[i])
```

Converting the targets into three numpy arrays for training, validation, and test sets.

```python
train, dev, test = DATA[1:53900], DATA[53900:65540], DATA[65540:]
RFtrain, RFtest = DATA[1:65000], DATA[65000:]
train_y, dev_y, test_y = np.zeros(len(train)), np.zeros(len(dev)), np.zeros(len(test))
RFtrain_y, RFtest_y = np.zeros(len(RFtrain)), np.zeros(len(RFtest))

for i in range(len(train)):
	try: train_y[i] = float(train[i][0]) 
	except: train_y[i] = 0.0
for i in range(len(dev)):
	try: dev_y[i] = float(dev[i][0]) 
	except: dev_y[i] = 0.0
for i in range(len(test)):
	try: test_y[i] = float(test[i][0]) 
	except: test_y[i] = 0.0
for i in range(len(RFtrain)):
	try: RFtrain_y[i] = float(RFtrain[i][0]) 
	except: RFtrain_y[i] = 0.0
for i in range(len(RFtest)):
	try: RFtest_y[i] = float(RFtest[i][0]) 
	except: RFtest_y[i] = 0.0

for i in range(5):
	print(train_y[i])
```

##### ***Data visualization***

```python
import matplotlib.pyplot as plt

step_size = 1000
plt.plot(range(0, len(train_y), step_size), train_y[::step_size], label = 'Train Data')
plt.plot(range(0, len(dev_y), step_size), dev_y[::step_size], label = 'Validation Data')
plt.plot(range(0, len(test_y), step_size), test_y[::step_size], label = 'Test Data')

plt.xlabel('Entry Number')
plt.ylabel('Temperature')
plt.title('DATA')
plt.legend()

plt.show()
```

#### ***Model Selection*** <br>
<br>
Initially we've selected a simple sequential network with all the layers being dense.<br>
After some experimentation with different hyperparameters these hyperparameters seemed to be best.

##### sequential model with dense layers

```python
model = tf.keras.Sequential([
	tf.keras.layers.Input(shape = (42,), batch_size = 100),
	tf.keras.layers.Dense(units = 128, activation = 'relu', kernel_regularizer = tf.keras.regularizers.l2(1E-3)),
	tf.keras.layers.Dense(units = 64, activation = 'relu', kernel_regularizer = tf.keras.regularizers.l2(1E-3)),
	tf.keras.layers.Dense(units = 1, activation = 'linear')
])

model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 1E-4), loss = tf.keras.losses.MeanSquaredError())

model.summary()
```

We've trained the model for 1000 epochs, using variable learning rate, adam optimizer.<br>
Training will stop if validation loss starts to plateau.

```python
epochs = 1000
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor = 'val_loss', factor = 0.6, patience = 20, min_lr = 1E-8)
early_stopping = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss', patience = 100, restore_best_weights = True)
history = model.fit(train_x, train_y, epochs = epochs, batch_size = 100, validation_data = (dev_x, dev_y), callbacks = [reduce_lr, early_stopping])

plt.plot(history.history['loss'][10:], label = 'Training Loss')
plt.plot(history.history['val_loss'][10:], label = 'Validation Loss')
plt.title('Loss Plot')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend()
```

Validation Loss obtained is poor. It starts to plateau near 30.

```python
model.save('./models/Dense.h5')
```

```python
model = tf.keras.models.load_model('./models/Dense.h5')
```

Let's calculate the test loss. It comes around 30.

```python
import pandas as pd

TestResults = model.predict(test_x)
TestResults = TestResults.flatten()

TestLoss = 0
for i in range(test_y.shape[0]):
	TestLoss += (test_y[i] - TestResults[i]) ** 2
TestLoss /= test_y.shape[0]
print("Loss in case of simple Dense layer: ", TestLoss)

TestPred = pd.DataFrame(TestResults)   
TestPred.to_csv('./files/TestPred.csv', index = False)   # saving the results and test data in csv files to compare manually
TestTrue = pd.DataFrame(test_y)
TestTrue.to_csv('./files/TestTrue.csv', index = False)
```

#### **Using LSTM Layers** 
LSTM works well for time series forecasting problems.<br>
Let's use stack two LSTM layers on top of each other followed by dense layers <br>
Since LSTM layer used input dimensions of type **[batch_size, timesteps, features]** we'll reshape the dataset into **[batch_size, 6, 7]**

```python
train_x = np.reshape(train_x, (train_x.shape[0], 6, 7))
dev_x = np.reshape(dev_x, (dev_x.shape[0], 6, 7))
test_x = np.reshape(test_x, (test_x.shape[0], 6, 7))
```

##### Model using LSTM layers

```python
modelLSTM = tf.keras.Sequential([
	tf.keras.Input(shape = (6, 7), batch_size = 100),
	tf.keras.layers.LSTM(units = 256, kernel_regularizer = tf.keras.regularizers.l2(0.005), return_sequences = True),
	tf.keras.layers.LSTM(units = 128, kernel_regularizer = tf.keras.regularizers.l2(0.005)),
	tf.keras.layers.Dense(units = 128, kernel_regularizer = tf.keras.regularizers.l2(0.005)),
	tf.keras.layers.Dense(units = 32, kernel_regularizer = tf.keras.regularizers.l2(0.005)),
	tf.keras.layers.Dense(units = 1, activation = 'linear')
])

modelLSTM.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 1E-4), loss = tf.keras.losses.MeanSquaredError())

modelLSTM.summary()
```

Training the model with the specified hyperparameters. After experimenting with several, learning rates, callbacks and layers' number and units. We came out with these hyperparameters. And it seems to give better result then previous one.

```python
epochs = 1000
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor = 'val_loss', factor = 0.6, patience = 20, min_lr = 1E-8)
early_stopping = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss', patience = 100, restore_best_weights = True)
history = modelLSTM.fit(train_x, train_y, epochs = epochs, batch_size = 100, validation_data = (dev_x, dev_y), callbacks = [reduce_lr, early_stopping])

plt.plot(history.history['loss'][10:], label = 'Training Loss')
plt.plot(history.history['val_loss'][10:], label = 'Validation Loss')
plt.title('Loss Plot')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend()
```

LSTM layers seem to improve the training accuracy, but validation accuracy is still poor.

```python
modelLSTM.save('./models/lstm.h5')
```

```python
modelLSTM = tf.keras.models.load_model('./models/lstm.h5')
```

Test accuracy comes out to be around 21 which a improvement over previous model.

```python
import pandas as pd

TestResults = modelLSTM.predict(test_x)
TestResults = TestResults.flatten()

TestLoss = 0
for i in range(test_y.shape[0]):
	TestLoss += (test_y[i] - TestResults[i]) ** 2
TestLoss /= test_y.shape[0]
print("Loss in case of LSTM: ", TestLoss)

TestPred = pd.DataFrame(TestResults)
TestPred.to_csv('./files/TestPredLSTM.csv', index = False)
TestTrue = pd.DataFrame(test_y)
TestTrue.to_csv('./files/TestTrueLSTM.csv', index = False)
```

#### **Trying different Sklearn Regressors suited for time series forecasting**

##### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import joblib

model = RandomForestRegressor(n_estimators = 500, random_state = 1)
model.fit(RFtrain_x, RFtrain_y)
joblib.dump(model, './models/RF.pkl')
```
